````python
print("="*80)
print("🧹 PIPELINE DE FILTRAGEM DE DADOS (NÍVEL LABORATÓRIO)")
print("="*80)

# ===== CONFIGURAÇÕES DE FILTRAGEM =====
MIN_CHARS = 200
MAX_REPETITION_RATIO = 0.3
MIN_UNIQUE_WORDS = 20
DEDUP_THRESHOLD = 0.85  # MinHash similarity threshold

print("\n⚙️  Configurações de Filtragem:")
print(f"   Mínimo de caracteres: {MIN_CHARS}")
print(f"   Máximo de repetição: {MAX_REPETITION_RATIO * 100}%")
print(f"   Mínimo de palavras únicas: {MIN_UNIQUE_WORDS}")
print(f"   Threshold de duplicação: {DEDUP_THRESHOLD}")

# ===== FUNÇÕES DE FILTRAGEM =====

def has_too_many_repetitions(text):
    """Detecta textos com repetições excessivas."""
    words = text.split()
    if len(words) < 50:
        return True
    
    counts = Counter(words)
    most_common_ratio = counts.most_common(1)[0][1] / len(words)
    
    return most_common_ratio > MAX_REPETITION_RATIO


def has_enough_unique_words(text):
    """Verifica se o texto tem palavras únicas suficientes."""
    words = text.split()
    unique_words = set(words)
    return len(unique_words) >= MIN_UNIQUE_WORDS


def is_valid_text(text):
    """
    Filtro principal de qualidade.
    Retorna True se o texto passar em todos os critérios.
    """
    if not text:
        return False
    
    text = text.strip()
    
    # Filtro 1: Comprimento mínimo
    if len(text) < MIN_CHARS:
        return False
    
    # Filtro 2: Repetições excessivas
    if has_too_many_repetitions(text):
        return False
    
    # Filtro 3: Diversidade de vocabulário
    if not has_enough_unique_words(text):
        return False
    
    # Filtro 4: Caracteres inválidos (aceitar português)
    if re.search(r"[^\x00-\x7FÀ-ÿ0-9.,;:!?()\"\'\-\s]", text):
        return False
    
    # Filtro 5: Detecção de idioma
    try:
        lang = detect(text)
        if lang != "pt":  # Foco em português
            return False
    except:
        return False
    
    return True


def get_minhash(text, num_perm=128):
    """Cria MinHash para deduplicação semântica."""
    m = MinHash(num_perm=num_perm)
    words = text.lower().split()
    for word in words:
        m.update(word.encode('utf8'))
    return m


def filter_and_deduplicate_texts(texts):
    """
    Pipeline completo de filtragem e deduplicação.
    Retorna lista de textos filtrados e deduplicated.
    """
    print("\n📊 Iniciando pipeline de filtragem...")
    
    # Etapa 1: Filtro de qualidade
    print("   🔍 Etapa 1: Aplicando filtros de qualidade...")
    filtered_texts = []
    
    for text in tqdm(texts, desc="   Filtrando"):
        if is_valid_text(text):
            filtered_texts.append(text)
    
    print(f"   ✅ Textos após filtro de qualidade: {len(filtered_texts):,} (removidos: {len(texts) - len(filtered_texts):,})")
    
    # Etapa 2: Deduplicação com MinHash
    print("\n   🔍 Etapa 2: Deduplicação avançada (MinHash LSH)...")
    lsh = MinHashLSH(threshold=DEDUP_THRESHOLD, num_perm=128)
    deduplicated_texts = []
    duplicates_found = 0
    
    for idx, text in enumerate(tqdm(filtered_texts, desc="   Deduplicando")):
        m = get_minhash(text)
        
        # Verificar se é similar a algum texto já adicionado
        if lsh.query(m):
            duplicates_found += 1
            continue
        
        # Adicionar ao índice e à lista final
        lsh.insert(f"text_{idx}", m)
        deduplicated_texts.append(text)
    
    print(f"   ✅ Duplicados removidos: {duplicates_found:,}")
    print(f"   ✅ Textos finais: {len(deduplicated_texts):,}")
    
    return deduplicated_texts


# ===== CRIAR DADOS BRUTOS (EXEMPLO) =====
print("\n📝 Criando dados de exemplo para demonstração...")

raw_texts = [
    "Este é um texto válido em português com informação relevante sobre machine learning.",
    "a a a a a a a a a a",  # Será removido: muita repetição
    "ML",  # Será removido: muito curto
    "This is an English text that should be filtered out.",  # Será removido: não é português
    "Alici foi criada por Mateus Nascimento dos Santos com mentalidade de fundador e visão de longo prazo.",
    "Alici foi criada por Mateus Nascimento dos Santos com mentalidade de fundador e visão de longo prazo.",  # Duplicata
    "O treinamento de modelos foundation requer dados de qualidade premium e filtrados adequadamente.",
    "♣♠♥♦ texto com caracteres inválidos",  # Será removido: caracteres inválidos
    "Inteligência artificial proprietária com identidade única desenvolvida com 1 bilhão de parâmetros.",
] * 200  # Simular dataset grande

print(f"   Dataset bruto criado: {len(raw_texts):,} textos")

# ===== EXECUTAR PIPELINE =====
clean_texts = filter_and_deduplicate_texts(raw_texts)

# ===== SALVAR DADOS LIMPOS =====
print("\n💾 Salvando dados limpos...")

with open("/content/datasets_clean/filtered_texts.txt", "w", encoding="utf-8") as f:
    for text in clean_texts:
        f.write(text + "\n\n")

print(f"✅ Salvo em: /content/datasets_clean/filtered_texts.txt")

# ===== ESTATÍSTICAS FINAIS =====
print("\n" + "="*80)
print("📊 ESTATÍSTICAS DE FILTRAGEM")
print("="*80)
print(f"   📥 Textos originais: {len(raw_texts):,}")
print(f"   ✅ Textos aprovados: {len(clean_texts):,}")
print(f"   ❌ Textos removidos: {len(raw_texts) - len(clean_texts):,}")
print(f"   📈 Taxa de aprovação: {(len(clean_texts) / len(raw_texts)) * 100:.1f}%")

avg_length = sum(len(t) for t in clean_texts) / len(clean_texts) if clean_texts else 0
print(f"\n   📏 Comprimento médio: {avg_length:.0f} caracteres")
print(f"   📏 Texto mais curto: {min(len(t) for t in clean_texts) if clean_texts else 0}")
print(f"   📏 Texto mais longo: {max(len(t) for t in clean_texts) if clean_texts else 0}")

print("\n" + "="*80)
print("✅ PIPELINE DE FILTRAGEM CONCLUÍDO!")
print("="*80)
print("\n💡 Mentalidade de Laboratório:")
print("   • Dataset bruto é VENENO")
print("   • Qualidade > Quantidade")
print("   • Seu maior ativo é o pipeline de dados")
print("   • Laboratórios sérios descartam 60-90% dos dados")
print("="*80)
````

````python
print("="*80)
print("🔥 TREINAMENTO DO MODELO FOUNDATION - ALICI (1B)")
print("="*80)

# === PREPARAR DATASET PARA TREINAMENTO ===
print("\n📊 Preparando dataset para treinamento...")

# Carregar textos premium
training_texts = []
with open("/content/datasets_clean/premium_texts.txt", "r", encoding="utf-8") as f:
    training_texts = [line.strip() for line in f if line.strip()]

# Carregar dados de identidade
with open("/content/datasets/alici_creator_identity.json", "r") as f:
    identity_data = json.load(f)
    training_texts.extend(identity_data["training_texts"])

# Carregar dados multimodais
with open("/content/datasets/alici_multimodal.json", "r") as f:
    multimodal_data = json.load(f)
    training_texts.extend([item["text"] for item in multimodal_data])

print(f"✅ Total de textos de treinamento: {len(training_texts):,}")

# Criar dataset do Hugging Face
class TextDatasetAlici(Dataset):
    def __init__(self, texts, tokenizer, max_length=1024):
        self.tokenizer = tokenizer
        self.texts = texts
        self.max_length = max_length
    
    def __len__(self):
        return len(self.texts)
    
    def __getitem__(self, idx):
        text = self.texts[idx]
        encodings = self.tokenizer(
            text,
            max_length=self.max_length,
            padding="max_length",
            truncation=True,
            return_tensors="pt"
        )
        return {
            "input_ids": encodings["input_ids"].squeeze(),
            "attention_mask": encodings["attention_mask"].squeeze()
        }

# Carregar tokenizer
tokenizer_path = "/content/tokenizer_alici/alici_tokenizer.json"
from transformers import PreTrainedTokenizerFast
custom_tokenizer = PreTrainedTokenizerFast(
    tokenizer_file=tokenizer_path,
    bos_token="<bos>",
    eos_token="<eos>",
    unk_token="<unk>",
    pad_token="<pad>"
)

# Criar dataset
train_dataset = TextDatasetAlici(training_texts, custom_tokenizer)

print(f"✅ Dataset criado: {len(train_dataset)} amostras")

# === CONFIGURAR TRAINER COM PARÂMETROS OTIMIZADOS PARA 1B ===
print("\n⚙️  Configurando Trainer para modelo de 1B...")

training_args = TrainingArguments(
    output_dir="/content/checkpoints",
    overwrite_output_dir=True,
    num_train_epochs=TRAINING_CONFIG['epochs'],
    per_device_train_batch_size=TRAINING_CONFIG['batch_size'],  # 16 para 1B
    save_steps=100,
    save_total_limit=2,
    logging_steps=20,
    weight_decay=TRAINING_CONFIG['weight_decay'],
    warmup_steps=TRAINING_CONFIG['warmup_steps'],  # 1000 para 1B
    learning_rate=TRAINING_CONFIG['learning_rate'],  # 3e-4 para 1B
    fp16=TRAINING_CONFIG['fp16'] if torch.cuda.is_available() else False,
    gradient_accumulation_steps=TRAINING_CONFIG['gradient_accumulation_steps'],  # 2 para compensar batch menor
    gradient_checkpointing=TRAINING_CONFIG['gradient_checkpointing'],
    optim="adamw_8bit" if torch.cuda.is_available() else "adamw_torch",
    max_grad_norm=1.0,
    seed=42,
    logging_first_step=True,
    report_to="none",  # Desabilitar wandb/tensorboard
)

trainer = Trainer(
    model=model,
    args=training_args,
    data_collator=DataCollatorForLanguageModeling(custom_tokenizer, mlm=False),
    train_dataset=train_dataset,
)

print("✅ Trainer configurado para 1B com:")
print(f"   Epochs: {training_args.num_train_epochs}")
print(f"   Batch Size: {training_args.per_device_train_batch_size}")
print(f"   Gradient Accumulation: {training_args.gradient_accumulation_steps}")
print(f"   Effective Batch: {training_args.per_device_train_batch_size * training_args.gradient_accumulation_steps}")
print(f"   Learning Rate: {training_args.learning_rate}")
print(f"   Warmup Steps: {training_args.warmup_steps}")
print(f"   FP16: {training_args.fp16}")
print(f"   Gradient Checkpointing: {training_args.gradient_checkpointing}")
print(f"   Optimizer: {training_args.optim}")

# === EXECUTAR TREINAMENTO ===
print("\n🚀 Iniciando treinamento do modelo de 1B...")
print("   ⚠️  Isto pode levar 10-20 minutos com GPU T4...")
print("   💡 Monitorando uso de memória GPU...")
print("-"*80)

# Mostrar uso de memória inicial
if torch.cuda.is_available():
    print(f"   GPU Memória Alocada: {torch.cuda.memory_allocated() / 1e9:.2f}GB")
    print(f"   GPU Memória Reservada: {torch.cuda.memory_reserved() / 1e9:.2f}GB")

try:
    train_result = trainer.train()
    
    print("-"*80)
    print(f"\n✅ TREINAMENTO CONCLUÍDO COM SUCESSO!")
    print(f"   Loss Final: {train_result.training_loss:.4f}")
    print(f"   Steps Total: {train_result.global_step}")
    
    if torch.cuda.is_available():
        print(f"   GPU Memória Final: {torch.cuda.memory_allocated() / 1e9:.2f}GB")

except RuntimeError as e:
    if "out of memory" in str(e):
        print("\n❌ ERRO: GPU sem memória suficiente!")
        print("   💡 Soluções:")
        print("      1. Reduza batch_size para 8 ou 4")
        print("      2. Aumente gradient_accumulation_steps")
        print("      3. Use GPU com mais VRAM (A100, V100)")
        print("      4. Reduza n_layer de 32 para 24")
        raise
    else:
        raise

print("\n" + "="*80)
print("✅ Modelo de 1B treinado e salvo em /content/checkpoints/")
print("="*80)
````
