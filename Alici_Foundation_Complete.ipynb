## 📦 Passo 1: Verificar e Instalar Dependências

````python
# Verificar versão do Python
import sys
print(f"Python version: {sys.version}")
print(f"Python executable: {sys.executable}")
````

````python
# Tentar importar dependências principais
print("Verificando dependências...\n")

dependencias = [
    ("fastapi", "FastAPI"),
    ("uvicorn", "Uvicorn"),
    ("psycopg2", "PostgreSQL driver"),
    ("tensorflow", "TensorFlow"),
    ("requests", "Requests"),
    ("dotenv", "Python-dotenv"),
    ("jose", "Python-JOSE (JWT)"),
    ("passlib", "Passlib"),
]

missing = []

for modulo, desc in dependencias:
    try:
        __import__(modulo)
        print(f"✅ {desc}")
    except ImportError:
        print(f"❌ {desc} - NÃO INSTALADO")
        missing.append(modulo)

if missing:
    print(f"\n⚠️  Módulos faltando: {', '.join(missing)}")
    print("\nInstale com: pip install " + " ".join(missing))
else:
    print("\n🎉 Todas as dependências estão instaladas!")
````

## 🗄️ Passo 2: Configurar Variáveis de Ambiente

````python
import os
from dotenv import load_dotenv

# Carregar .env
load_dotenv()

DATABASE_URL = os.getenv("DATABASE_URL")

if DATABASE_URL:
    # Ocultar senha na exibição
    url_safe = DATABASE_URL[:30] + "..." if len(DATABASE_URL) > 30 else DATABASE_URL
    print(f"✅ DATABASE_URL configurada: {url_safe}")
else:
    print("❌ DATABASE_URL não encontrada!")
    print("Configure no arquivo .env")
    print("\nExemplo:")
    print("DATABASE_URL=postgresql://user:pass@host.neon.tech/alici?sslmode=require")
````

## 🔨 Passo 3: Inicializar Banco de Dados

````python
# Executar init_db.py
print("Inicializando banco de dados...\n")

try:
    from init_db import criar_tabelas, verificar_conexao
    
    # Verificar conexão
    if verificar_conexao():
        print("\n✅ Conexão com banco estabelecida")
        
        # Criar tabelas
        print("\nCriando tabelas...")
        if criar_tabelas():
            print("\n✅ Banco de dados inicializado com sucesso!")
        else:
            print("\n❌ Erro ao criar tabelas")
    else:
        print("\n❌ Não foi possível conectar ao banco")
        
except Exception as e:
    print(f"❌ Erro: {e}")
````

## 🧠 Passo 4: Testar Engine de IA (6 Camadas)

````python
# Testar importação do engine
print("Testando engine ALICI...\n")

try:
    from engine import gerar_resposta
    
    print("✅ Engine importado com sucesso\n")
    
    # Testes das 6 camadas
    testes = [
        ("quem é você", "Camada 1: Identidade"),
        ("oi", "Camada 3: Regras Locais"),
        ("como você está", "Camada 3: Regras Locais"),
    ]
    
    for pergunta, camada in testes:
        print(f"📝 Testando [{camada}]: '{pergunta}'")
        resposta = gerar_resposta(pergunta)
        preview = resposta[:80] + "..." if len(resposta) > 80 else resposta
        print(f"✅ Resposta: {preview}\n")
    
    print("🎉 Engine funcionando corretamente!")
    
except Exception as e:
    print(f"❌ Erro ao testar engine: {e}")
````

## 📊 Passo 5: Gerar Dataset de Treinamento

````python
# Executar geração de dataset
print("Gerando dataset...\n")

try:
    from gerar_dataset import gerar_dataset_base, adicionar_variacoes, adicionar_perguntas_avancadas, salvar_dataset, gerar_estatisticas
    
    # Gerar dataset
    dataset = gerar_dataset_base()
    print(f"✅ {len(dataset)} pares base criados")
    
    # Adicionar variações
    variacoes = adicionar_variacoes(dataset)
    dataset.extend(variacoes)
    print(f"✅ {len(variacoes)} variações adicionadas")
    
    # Adicionar perguntas avançadas
    avancadas = adicionar_perguntas_avancadas()
    dataset.extend(avancadas)
    print(f"✅ {len(avancadas)} perguntas avançadas adicionadas")
    
    # Salvar
    total = salvar_dataset(dataset)
    print(f"\n✅ Dataset salvo: {total} pares únicos")
    
    # Estatísticas
    stats = gerar_estatisticas(dataset)
    print(f"\n📊 Estatísticas:")
    print(f"  • Total: {stats['total_pares']}")
    print(f"  • Média pergunta: {stats['media_len_pergunta']:.1f} caracteres")
    print(f"  • Média resposta: {stats['media_len_resposta']:.1f} caracteres")
    
except Exception as e:
    print(f"❌ Erro: {e}")
````

## 🤖 Passo 6: Verificar Modelos de ML

````python
# Verificar TensorFlow e modelos
print("Verificando modelos ML...\n")

try:
    import tensorflow as tf
    print(f"✅ TensorFlow {tf.__version__} instalado")
    
    # Listar modelos .h5
    import os
    print("\nBuscando modelos .h5...")
    
    locais = [".", "model", "Modelo"]
    modelos_encontrados = []
    
    for local in locais:
        if os.path.exists(local):
            for arquivo in os.listdir(local):
                if arquivo.endswith(".h5"):
                    caminho = os.path.join(local, arquivo)
                    tamanho = os.path.getsize(caminho) / (1024 * 1024)  # MB
                    modelos_encontrados.append((caminho, tamanho))
                    print(f"  ✅ {caminho} ({tamanho:.2f} MB)")
    
    if not modelos_encontrados:
        print("  ⚠️  Nenhum modelo .h5 encontrado")
        print("\n  Para treinar modelos:")
        print("  1. Execute as células acima para gerar dataset")
        print("  2. Use colab_finetuning.py no Google Colab")
        print("  3. Baixe o modelo treinado para pasta model/")
    else:
        print(f"\n✅ {len(modelos_encontrados)} modelo(s) encontrado(s)")
        
except Exception as e:
    print(f"❌ Erro: {e}")
````

````python
# Executar verificação completa
print("Executando verificação completa do sistema...\n")
print("=" * 70)

try:
    from init_alici import (
        verificar_dependencias,
        verificar_arquivos_core,
        verificar_configuracao,
        verificar_banco_dados,
        verificar_engine,
        verificar_modelos,
        verificar_templates
    )
    
    resultados = {
        "Dependências": verificar_dependencias(),
        "Arquivos Core": verificar_arquivos_core(),
        "Configuração": verificar_configuracao(),
        "Banco de Dados": verificar_banco_dados(),
        "Engine IA": verificar_engine(),
        "Modelos ML": verificar_modelos(),
        "Templates": verificar_templates(),
    }
    
    print("\n" + "=" * 70)
    print("📊 RESUMO DA VERIFICAÇÃO")
    print("=" * 70)
    
    for componente, status in resultados.items():
        simbolo = "✅" if status else "❌"
        print(f"{simbolo} {componente}")
    
    print()
    
    # Status geral
    total = len(resultados)
    aprovados = sum(resultados.values())
    porcentagem = (aprovados / total) * 100
    
    print(f"Status Geral: {aprovados}/{total} componentes OK ({porcentagem:.1f}%)")
    
    if porcentagem == 100:
        print("\n🎉 Sistema 100% operacional!")
    elif porcentagem >= 70:
        print("\n⚠️  Sistema parcialmente operacional")
    else:
        print("\n❌ Sistema não está pronto")
        
except Exception as e:
    print(f"❌ Erro na verificação: {e}")
````

## 🧪 Passo 8: Teste Completo do Engine

````python
# Executar suite de testes completa
print("Executando testes do engine...\n")
print("=" * 70)

try:
    from teste_engine_completo import (
        testar_importacoes,
        testar_identidade,
        testar_regras_locais,
        testar_aprendizado,
        testar_intencao_web,
        testar_fallback,
        testar_resposta_com_emocao,
        testar_pipeline_completo
    )
    
    testes = [
        ("Importações", testar_importacoes),
        ("Identidade", testar_identidade),
        ("Regras Locais", testar_regras_locais),
        ("Aprendizado", testar_aprendizado),
        ("Intenção Web", testar_intencao_web),
        ("Fallback", testar_fallback),
        ("Sistema de Emoções", testar_resposta_com_emocao),
        ("Pipeline Completo", testar_pipeline_completo),
    ]
    
    resultados = {}
    
    for nome, teste_func in testes:
        try:
            resultados[nome] = teste_func()
        except Exception as e:
            print(f"\n  ❌ Erro ao executar teste '{nome}': {e}")
            resultados[nome] = False
    
    # Resumo
    print("\n" + "=" * 70)
    print("📊 RESUMO DOS TESTES")
    print("=" * 70)
    
    for nome, passou in resultados.items():
        simbolo = "✅" if passou else "❌"
        print(f"{simbolo} {nome}")
    
    print()
    
    total = len(resultados)
    aprovados = sum(resultados.values())
    porcentagem = (aprovados / total) * 100
    
    print(f"Resultado: {aprovados}/{total} testes passaram ({porcentagem:.1f}%)")
    
    if porcentagem == 100:
        print("\n🎉 Todos os testes passaram! Engine 100% operacional!")
    elif porcentagem >= 75:
        print("\n⚠️  Maioria dos testes passou. Sistema funcional.")
    else:
        print("\n❌ Muitos testes falharam. Verifique os erros acima.")
        
except Exception as e:
    print(f"❌ Erro ao executar testes: {e}")
````
